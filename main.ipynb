```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from functions import load_and_clean_csv
```

```python
url = "amz_uk_price_prediction_dataset.csv"
df = load_and_clean_csv(url)
df.head()
```

Part 1: Understanding Product Categories
1. Frequency Tables

To begin, let's generate a frequency table for the category column to determine the most popular product categories.

```python
# Frequency table for product categories
category_frequency = df['category'].value_counts()

# Display the top 5 most listed product categories
top_5_categories = category_frequency.head(5)
print(top_5_categories)
```

2. Visualizations

Bar Chart: To visualize the distribution of products across different categories.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=category_frequency.index[:10], y=category_frequency.values[:10])
plt.xticks(rotation=90)
plt.title('Top 10 Most Listed Product Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

Pie Chart: Visualizing the proportions of the top categories.

```python
plt.figure(figsize=(7, 7))
plt.pie(top_5_categories, labels=top_5_categories.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Top 5 Product Categories')
plt.show()
```

Part 2: Delving into Product Pricing
1. Measures of Centrality

To understand the central tendencies of product pricing:

```python
# Measures of centrality for product prices
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]  # Mode might return multiple values; take the first

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

2. Measures of Dispersion

To measure the variation in product prices:

```python
# Measures of dispersion for product prices
variance_price = df['price'].var()
std_deviation_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_deviation_price}")
print(f"Range: {range_price}")
print(f"Interquartile Range: {iqr_price}")
```

3. Visualizations

Histogram: This will help you see the distribution of product prices and identify if most prices fall within a particular range.

```python
# Histogram of product prices
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Box Plot: This will help detect outliers and the spread of prices.


```python
# Boxplot of product prices
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of Product Prices')
plt.show()
```

Part 3: Unpacking Product Ratings
1. Measures of Centrality

For ratings, you can calculate the mean, median, and mode:

```python
# Measures of centrality for product ratings
mean_stars = df['stars'].mean()
median_stars = df['stars'].median()
mode_stars = df['stars'].mode()[0]  # Take the first mode value

print(f"Mean stars: {mean_stars}")
print(f"Median stars: {median_stars}")
print(f"Mode stars: {mode_stars}")
```

2. Measures of Dispersion

To assess how spread out the ratings are:

```python
# Measures of dispersion for product ratings
variance_stars= df['stars'].var()
std_deviation_stars = df['stars'].std()
iqr_stars = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Variance: {variance_stars}")
print(f"Standard Deviation: {std_deviation_stars}")
print(f"Interquartile Range: {iqr_stars}")
```

3. Shape of the Distribution

Understanding whether ratings are skewed and how peaked the distribution is:

```python
# Skewness and Kurtosis of product ratings
skewness_stars = df['stars'].skew()
kurtosis_stars = df['stars'].kurtosis()

print(f"Skewness: {skewness_stars}")
print(f"Kurtosis: {kurtosis_stars}")
```

4. Visualizations

Histogram: To visualize the distribution of ratings.

```python
# Histogram of product ratings
plt.figure(figsize=(10, 6))
plt.hist(df['stars'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('stars')
plt.ylabel('Frequency')
plt.show()
```
